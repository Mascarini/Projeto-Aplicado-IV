# **Introdução**

A dengue continua sendo um dos principais desafios de saúde pública no Brasil e apresenta comportamento cíclico com períodos de alta e queda ao longo do ano. No estado de São Paulo, esse cenário é ainda mais relevante pela grande heterogeneidade entre municípios, além da influência de fatores climáticos e demográficos na dinâmica de transmissão. Nesse contexto, o projeto mantém o recorte temporal de 2015 a 2026 para analisar a evolução da incidência da doença de forma estruturada.

Com base no que foi delimitado na entrega 1, esta entrega 2 consolida o direcionamento técnico da solução. A motivação central é transformar dados públicos em informação útil para tomada de decisão, com foco em prevenção e resposta mais rápida em períodos críticos. O objetivo geral é modelar o comportamento temporal da dengue em São Paulo para apoiar previsões de curto prazo. Como objetivos específicos, busca-se: (1) organizar e padronizar os dados, (2) identificar tendências, sazonalidade e anomalias, (3) comparar abordagens de modelagem temporal e (4) traduzir os resultados em análises interpretáveis para uso aplicado.

As bases utilizadas são os arquivos `dataset/infodengue_dengue_sp_state.csv` e `dataset/infodengue_dengue_sp_municipalities.csv`, provenientes do ecossistema InfoDengue/Mosqlimate. A justificativa do projeto combina relevância social, disponibilidade de dados em série temporal e potencial prático de geração de alertas e planejamento de ações em saúde pública.

# **Referencial Teórico**

A análise de séries temporais é amplamente utilizada em epidemiologia para compreender como doenças evoluem ao longo do tempo. Em cenários como dengue, os componentes de tendência, sazonalidade e variação residual são fundamentais para explicar períodos de maior risco e apoiar decisões preventivas (OPAS, s.d.; WHO, 2025). A interpretação desses componentes ajuda a distinguir aumentos esperados (sazonais) de desvios relevantes (anomalias), o que é essencial para monitoramento em saúde pública (WHO, 2025).

No contexto deste projeto, os dados do InfoDengue permitem acompanhar a evolução semanal dos casos e incorporar variáveis contextuais, como condições climáticas, incidência e nível de alerta (FIOCRUZ, s.d.). Além disso, a literatura institucional destaca que ações de vigilância epidemiológica combinadas com análise de dados são importantes para antecipar picos de transmissão e orientar medidas de prevenção (OPAS, s.d.; BRASIL, s.d.). A qualidade do pré-processamento, a consistência temporal e a escolha de métricas de avaliação têm impacto direto na utilidade prática dos modelos.

Como alternativas de solução, há abordagens estatísticas clássicas (como decomposição temporal e modelos autoregressivos) e abordagens de aprendizado de máquina. Métodos clássicos tendem a oferecer maior interpretabilidade e transparência, enquanto métodos de ML podem capturar relações mais complexas entre variáveis quando há maior volume e qualidade de dados. Em contrapartida, técnicas mais complexas podem exigir mais ajuste, maior custo computacional e cuidado adicional para evitar sobreajuste.

Dessa forma, a estratégia adotada nesta entrega é comparar abordagens de forma incremental, iniciando por modelos baseline e avançando para métodos mais sofisticados quando houver ganho real em desempenho. O critério de escolha prioriza equilíbrio entre acurácia, robustez e capacidade de explicação dos resultados para o contexto de saúde pública, conforme recomendações de monitoramento e resposta de órgãos de saúde (WHO, 2025; OPAS, s.d.).

# **Pipeline da Solução**

1. **Ingestão e checagem inicial dos dados**
   - Ler `dataset/infodengue_dengue_sp_state.csv` e `dataset/infodengue_dengue_sp_municipalities.csv`.
   - Validar colunas obrigatórias, tipos de dados e consistência do período 2015-2026.
   - Verificar valores faltantes, duplicidades e possíveis inconsistências de registro.

2. **Pré-processamento temporal e engenharia de atributos**
   - Padronizar datas e ordenar as séries por semana epidemiológica.
   - Tratar ausentes e possíveis outliers com critério epidemiológico.
   - Criar atributos temporais (mês, trimestre, sazonalidade, defasagens e médias móveis).
   - Definir recortes estadual e municipal para comparações analíticas.

3. **Análise exploratória e diagnóstico da série**
   - Analisar tendência, sazonalidade e variabilidade da incidência.
   - Identificar semanas de pico e possíveis períodos anômalos.
   - Comparar comportamento entre municípios de maior incidência e o agregado estadual.

4. **Modelagem preditiva e validação**
   - Construir baseline de referência para comparação.
   - Treinar modelos temporais com divisão respeitando ordem cronológica.
   - Avaliar desempenho com MAE, RMSE e MAPE, além de inspeção visual de resíduos.
   - Selecionar o modelo com melhor equilíbrio entre desempenho e interpretabilidade.

5. **Comunicação dos resultados e aplicação prática**
   - Construir visualizações objetivas para público técnico e não técnico.
   - Traduzir achados em recomendações para planejamento preventivo.
