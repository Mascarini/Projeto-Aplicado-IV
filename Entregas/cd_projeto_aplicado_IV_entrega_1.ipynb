# **Introdução**

A dengue é uma doença transmitida por mosquito que há anos representa um grande desafio para a saúde pública no Brasil. Sua transmissão ocorre de forma cíclica, com períodos de aumento expressivo de casos, especialmente em determinadas épocas do ano. No estado de São Paulo, o crescimento urbano acelerado, aliado às variações climáticas mais intensas registradas na última década, tornou necessário um acompanhamento constante da situação (FIOCRUZ, s.d.).

Este projeto propõe analisar como a doença evoluiu ao longo de 11 anos, de 2015 até 2026. O foco está em entender como os casos se distribuíram nos municípios e de que forma fatores ambientais e possíveis mudanças no comportamento do vírus influenciaram esse cenário.

Ao reunir dados históricos e informações mais recentes de 2026, a intenção é avaliar se as medidas de controle adotadas ao longo dos anos trouxeram resultados efetivos ou se surgiram novos fatores que passaram a influenciar o aumento dos casos.
